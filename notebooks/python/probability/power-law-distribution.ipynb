# Power Law Distribution

A heavy-tailed distribution where:

$P(X = x) \propto x^{-\alpha}$

- Parameter: $\alpha > 1$
- Very common in natural networks, physics, economics.

**Characteristics:**

| Property               | Description                                                                |
| :--------------------- | :------------------------------------------------------------------------- |
| Tail behavior          | Heavy tail (slow decay) — large events have non-negligible probability     |
| Scale                  | Scale-free / no characteristic scale                                       |
| Typical $\alpha$ range | $1 < \alpha < 3$ in many real systems                                      |
| Moments                | If $\alpha \le 2$ mean may diverge; if $\alpha \le 3$ variance may diverge |
| Examples               | City populations, word frequencies, wealth, internet traffic               |

**Practical note:** estimating $\alpha$ reliably requires many samples from the tail. Methods: linear regression on log-log plot (biased), maximum likelihood estimation (preferred).

**Use Case & Example**:

| Scenario / Domain           | Description                                                                                              |
| :-------------------------- | :------------------------------------------------------------------------------------------------------- |
| **City Populations**        | A few cities (São Paulo, Tokyo, NYC) are extremely large, while most cities are much smaller.            |
| **Earthquake Magnitudes**   | Small earthquakes are very common; huge earthquakes are extremely rare but follow a power-law frequency. |
| **Wealth Distribution**     | A small number of people hold most of the wealth; many people have relatively little.                    |
| **Forest Fire Sizes**       | Many tiny fires occur each year, while very large fires happen rarely but disproportionately matter.     |
| **File Sizes on the Web**   | Most files are small, but a few extremely large files dominate bandwidth usage.                          |
| **Internet Traffic**        | A small fraction of websites (e.g., Google, YouTube) account for most traffic.                           |
| **Word Frequencies (Zipf)** | In any language, a few words occur extremely frequently (the, and, of), while most appear rarely.        |
| **Scientific Citations**    | A small number of papers receive very high citation counts; most receive few.                            |
| **Social Networks**         | Most users have just a few connections, but some users (influencers) have extremely many.                |
| **Solar Flare Intensities** | Many small flares occur; massive solar flares are rare but follow power-law behavior.                    |
