# Uniform Distribution (Continuous or Discrete)

All outcomes within a specified range ($a$ to $b$) are equally likely.

#### Continuous Uniform Distribution

- Function: PDF
- PDF Formula: $\frac{1}{b-a}$ for $X \in [a,b]$, otherwise 0.
- Use Case & Example: The time a bus takes to arrive is uniformly distributed between 10 and 20 minutes. The probability is $\frac{1}{20-10} = \frac{1}{10}$ for any given minute in that interval.

#### Discrete Uniform Distribution

- Function: PMF
- PMF Formula: $\frac{1}{n}$ (where $n$ is the total number of outcomes).
- Use Case & Example: Rolling a fair 6-sided die. The probability of any single outcome $\{1, 2, 3, 4, 5, 6\}$ is $\frac{1}{6}$.
