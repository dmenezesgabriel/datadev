# Singly Linked Lists


Define a node


```python
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next

    def __str__(self):
        return str(self.value)
```

Create a Singly Linked List


```python
head = Node(1)
a = Node(2)
b = Node(3)
c = Node(4)

head.next = a
a.next = b
b.next = c
```

Traverse the list - O(n)


```python
cursor = head
while cursor:
    print(cursor)
    cursor = cursor.next
```

Display the linked list - O(n)


```python
def representation(head):
    if head is None:
        return "[]"
    else:
        last = head
        return_string = f"[{last.value}"

        while last.next:
            last = last.next
            return_string += f", {last.value}"
        return_string += "]"

        return return_string
```

Representation


```python
representation(head)
```
