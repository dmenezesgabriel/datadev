# Graphs

- Is composed by vertices and edges
- Trees and linked lists are a subclass of graphs

Types:

- **Directed graph**: There is a direction on the graph connections ->
- **Undirected graph**: Every edge is both "to" and "from" <->
- **Cyclic graph**: There is a cycle in it like a loop


```python
from collections import defaultdict
```

## Storing

**Edge List**:

- Directed graph: `[[From, To], [From, To]]`
- Undirected graph: `[[Each number represents both From and To]]`


```python
list_of_edges = [
    [0, 1],
    [1, 2],
    [3, 4],
    [3, 6],
    [3, 7],
    [4, 2],
    [4, 5],
    [5, 2],
]
```

**Adjacency Matrix**:

- List of lists: Connections are represented by 1

| 0   | 1   | 2   | 3   | 4   | 5   |
| --- | --- | --- | --- | --- | --- |
| 1   | 0   | 0   | 1   | 0   | 1   |
| 2   | 0   | 1   | 0   | 0   | -   |
| 3   | -   | -   | -   | -   | -   |
| 4   | -   | -   | -   | -   | -   |
| 5   | -   | -   | -   | -   | -   |


## Examples


```python
n = 8
edges = [
    [0, 1],
    [1, 2],
    [0, 3],
    [3, 4],
    [3, 6],
    [3, 7],
    [4, 2],
    [4, 5],
    [5, 2],
]
```

```python
matrix = []
for i in range(n):
    matrix.append([0] * n)
matrix
```

```python
for u, v in edges:
    matrix[u][v] = 1
    # matrix[v][u] = 1 if undirected graph

matrix
```

```python
default_dict = defaultdict(list)

for u, v in edges:
    default_dict[u].append(v)
    # default_dict[v].append(u) = 1 if undirected graph


default_dict
```
