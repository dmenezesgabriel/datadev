```python
def isPalindrome(s: str) -> bool:
    text = "".join(char.lower() for char in s if char.isalnum())
    left = 0
    right = len(text) - 1

    while left < right:
        if text[left] != text[right]:
            print(text[left])
            print(text[right])
            return False
        left += 1
        right -= 1

    return True
    
s = "A man, a plan, a canal: Panama"
isPalindrome(s)
```
