```python
from typing import List

def twoSum(numbers: List[int], target: int) -> List[int]:
    left = 0
    right = len(numbers) - 1
    
    
    while numbers[left] + numbers[right] != target:
        s = numbers[left] + numbers[right]
        if s > target:
            right -= 1
        else:
            left += 1
        
    return [left +1, right + 1]
    
    
    

numbers = [2,7,11,15]
target = 9
twoSum(numbers, target)
```
