# Power Law Distribution

A heavy-tailed distribution where:

$P(X = x) \propto x^{-\alpha}$

- Parameter: $\alpha > 1$
- Very common in natural networks, physics, economics.

**Characteristics:**

| Property               | Description                                                                |
| :--------------------- | :------------------------------------------------------------------------- |
| Tail behavior          | Heavy tail (slow decay) â€” large events have non-negligible probability     |
| Scale                  | Scale-free / no characteristic scale                                       |
| Typical $\alpha$ range | $1 < \alpha < 3$ in many real systems                                      |
| Moments                | If $\alpha \le 2$ mean may diverge; if $\alpha \le 3$ variance may diverge |
| Examples               | City populations, word frequencies, wealth, internet traffic               |

**Practical note:** estimating $\alpha$ reliably requires many samples from the tail. Methods: linear regression on log-log plot (biased), maximum likelihood estimation (preferred).
